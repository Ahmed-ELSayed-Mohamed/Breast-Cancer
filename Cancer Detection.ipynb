```python
#import dataset from sklearn datasets
cancer=load_breast_cancer()
#cancer
```

```python
#Convert dataset to Dataframe using Panda
data_set= pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
data_set['target']=cancer['target']
```

```python
# preview of data (5 first result)
data_set.head()
```

# step 2 Visualization the Data

```python
# shows the relation between the input data and the targer
sns.pairplot(data_set, hue = 'target', vars = ['mean radius','mean texture', 'mean area', 'mean perimeter', 'mean smoothness'])
```

```python

plt.figure(figsize=(20,10))
# compare all the features on a Vs. correlation
sns.heatmap(data_set.corr(),annot=True)
```

# step: 4 start traing model

```python
# drop the target column from the dataset to create the input of the model
X=data_set.drop('target',axis=1)
```

```python
# create the expected output
Y=data_set['target']
```

# now we will split or data into train data and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=.33,random_state=44,shuffle=True)
```

# we import SVM model to train our model                    

```python
from sklearn.svm import SVC
svc=SVC()
```

```python
svc.fit(x_train,y_train)
```

# Step : Evalute the Model 

```python
#predict the output of our model according using the testing dataset
y_predict=svc.predict(x_test)
```

```python
# calculate the confusion matrix in order to evaluate the results
from sklearn.metrics import classification_report ,confusion_matrix
cn_mat=confusion_matrix(y_test,y_predict)
cn_mat
```

```python
#show comparizon using Heat Map

sns.heatmap(cn_mat,annot=True)
```

```python
print(classification_report(y_test,y_predict))
```

# here when we use metrics to measure the accurcy of the model we find it so bad so we need to improve or model
